```
#ตัดcolumn ที่ไม่ใช้ออก ให้เหลือแต่ ชื่อของนักศึกษาเพื่อนำไปตัดกรด และ ชื่อกรรมการเพื่อหาไปถ่วงน้ำหนักคะแนน
preday1 = Data_midterm3_Day1[['นักศึกษา','กรรมการสอบ','รวม']]
preday2 = Data_midterm3_Day2[['นักศึกษา','กรรมการสอบ','รวม']]
preday3 = Data_midterm3_Day3[['นักศึกษา','กรรมการสอบ','รวม']]
```

```
#รวมคะแนนในแต่ละวันมาไว้
Scoresall = pd.concat([preday1,preday2,preday3])
Scoresall 
```

```
#คำนวนค่าเฉลี่ยของกรรมการคุมสอบ
mean_data3 = Scoresall.groupby('กรรมการสอบ').mean()
mean_data3
```

```
#คำนวนส่วนเบี่ยงเบนมาตรฐานของกรรมการคุมสอบ
std_data3 = Scoresall.groupby('กรรมการสอบ').std()
std_data3
```

```
#หา z score จากที่อาจารย์กำหนด
def cal_adjusted_score(z_score):
    return round(30 + ((-15*(z_score-2.3103448275862064))/(-4.788052880847054)),2)
  
```
